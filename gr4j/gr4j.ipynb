# Hydrological Modelling

Python implementation of the GR4J hydrological model, developped at [Irstea](http://www.irstea.fr/en/accueil) (see [this article](http://dx.doi.org/10.1016/S0022-1694%2803%2900225-7) or [this web site](http://webgr.irstea.fr/?lang=en)). You will need to have [Numba](http://numba.pydata.org/) installed, as well as Matplotlib, NumPy, SciPy and Pandas (I recommend using the [Anaconda](http://continuum.io/downloads.html) distribution).

First import the usual libraries.

```python
%matplotlib inline
import matplotlib.pyplot as plt
from numba import jit
import pandas as pd
from pandas import DataFrame, Series
import math
import numpy as np
import scipy.optimize
```

This is the GR4J hydrological model, implemented as a Python class and a Numba function for speeding up the simulation (notice the @jit decorator for Just In Time compilation).

```python
@jit
def run_gr4j(x, p, e, q, s, uh1_tab, uh2_tab, l, m):
    for t in range(p.size):
        if p[t] > e[t]:
            pn = p[t] - e[t]
            en = 0.
            tmp = s[0] / x[0]
            ps = x[0] * (1. - tmp * tmp) * math.tanh(pn / x[0]) / (1. + tmp * math.tanh(pn / x[0]))
            s[0] += ps
        elif p[t] < e[t]:
            ps = 0.
            pn = 0.
            en = e[t] - p[t]
            tmp = s[0] / x[0]
            es = s[0] * (2. - tmp) * np.tanh(en / x[0]) / (1. + (1. - tmp) * np.tanh(en / x[0]))
            tmp = s[0] - es
            if tmp > 0.:
                s[0] = tmp
            else:
                s[0] = 0.
        else:
            pn = 0.
            en = 0.
            ps = 0.
        tmp = (4. * s[0] / (9. * x[0]))
        perc = s[0] * (1. - (1. + tmp * tmp * tmp * tmp) ** (-1. / 4.))
        s[0] -= perc
        pr_0 = perc + pn - ps
        q9 = 0.
        q1 = 0.
        for i in range(m):
            if i == 0:
                pr_i = pr_0
            else:
                pr_i = s[2 + i - 1]
            if i < l:
                q9 += uh1_tab[i] * pr_i;
            q1 += uh2_tab[i] * pr_i;
        q9 *= 0.9
        q1 *= 0.1
        f = x[1] * ((s[1] / x[2]) ** (7. / 2.))
        tmp = s[1] + q9 + f
        if tmp > 0.:
            s[1] = tmp
        else:
            s[1] = 0.
        tmp = s[1] / x[2]
        qr = s[1] * (1. - ((1. + tmp * tmp * tmp * tmp) ** (-1. / 4.)))
        s[1] -= qr
        tmp = q1 + f
        if tmp > 0.:
            qd = tmp
        else:
            qd = 0.
        q[t] = qr + qd
        for i in range(s.size - 2 - 2, -1, -1):
            s[2 + i + 1] = s[2 + i]
        if s.size > 2:
            s[2] = pr_0

class gr4j:
    def sh1(self, t):
        if t == 0:
            res = 0.
        elif t < self.x[3]:
            res = (float(t) / self.x[3]) ** (5. / 2.)
        else:
            res = 1.
        return res
    def sh2(self, t):
        if t == 0:
            res = 0.
        elif t < self.x[3]:
            res = 0.5 * ((float(t) / self.x[3]) ** (5. / 2.))
        elif t < 2. * self.x[3]:
            res = 1. - 0.5 * ((2. - float(t) / self.x[3]) ** (5. / 2.))
        else:
            res = 1.
        return res
    def uh1(self, j):
        return self.sh1(j) - self.sh1(j - 1)
    def uh2(self, j):
        return self.sh2(j) - self.sh2(j - 1)
    def __init__(self, x, s = None):
        self.x = x
        if s == None:
            self.s = np.array([0., 0.] + [0.] * int(2. * self.x[3]))
        else:
            self.s = np.array(s)
        self.l = int(self.x[3]) + 1
        self.m = int(2. * self.x[3]) + 1
        self.uh1_tab = np.empty(self.l)
        self.uh2_tab = np.empty(self.m)
        for i in range(self.m):
            if i < self.l:
                self.uh1_tab[i] = self.uh1(i + 1)
            self.uh2_tab[i] = self.uh2(i + 1)
    def run(self, pe):
        q = Series(np.empty_like(pe.p), index = pe.index)
        run_gr4j(self.x, pe.p.values, pe.e.values, q.values, self.s, self.uh1_tab, self.uh2_tab, self.l, self.m)
        return q
    def set_conf(self, conf):
        if conf == 0:
            for i in range(self.s.size):
                self.s[i] = 0.
        elif conf == 1:
            self.s[0] = self.x[0] / 2.
            self.s[1] = self.x[2] / 2.
            for i in range(self.s.size - 2):
                self.s[i + 2] = 0.
```

Load some data. For this model we just need the precipitation (p), potential evapotranspiration (e) and streamflow (q) time series.

```python
peq = pd.read_pickle('peq.pkl')
plt.close('all')
plt.figure(figsize = (20, 10))
peq.p.plot(label = 'Precipitation')
peq.e.plot(label = 'Potential evapotranspiration')
peq.q.plot(label = 'Streamflow')
plt.ylabel('mm/day')
plt.legend()
plt.show()
```

The 4 parameters of the model will be calibrated using the observed precipitation, potential evapotranspiration and streamflow time series. The *calibration* function will be passed to an optimizing function. We will also need an error function, which measures how well the simulated streamflow fits the observed streamflow. Here we use the classic Root Mean Square Error.

```python
def calibration(x, model, peq, error_func, warmup_period = 0, x_ranges = None, init_conf = 0, trace = None):
    if x_ranges != None:
        for i, x_range in enumerate(x_ranges):
            if x[i] < x_range[0] or x[i] > x_range[1]:
                return np.inf
    q_mod = model(x)
    q_mod.set_conf(init_conf)
    q_sim = q_mod.run(peq)
    error = error_func(peq.q[warmup_period:], q_sim[warmup_period:])
    if trace != None:
        trace['x'].append(x)
        trace['error'].append(error)
    return error

def rmse(observations, simulations):
    diff = observations - simulations
    diff = diff[~np.isnan(diff)]
    return np.sqrt(np.mean(np.square(diff)))
```

Now let's find the optimal model parameters. We specify a starting point (*x0*), a warm-up period (here 365 days), some ranges for the parameters, and an initial model configuration (the catchment should not be completely dry!). The *optimize* function can take a bit of time, so be patient...

```python
x0 = [100., 0., 100., 1.]
warmup_period = 365
x_ranges = [[0., np.inf],
            [-np.inf, np.inf],
            [0., np.inf],
            [0., np.inf]]
init_conf = 1
trace = {'x': [], 'error': []}
x = scipy.optimize.fmin(calibration, x0, args = (gr4j, peq, rmse, warmup_period, x_ranges, init_conf, trace), xtol = 1e-3, disp = False)
```

We plot the traces of the optimizing function. Of course there is no guarantee that we found the best solution (it could be a local optimum).

```python
print('X = ' + str(x))
plt.close('all')
plt.figure(figsize = (20, 15))
for x_i in range(4):
    plt.subplot(511 + x_i)
    plt.plot([i[x_i] for i in trace['x']], label = 'X' + str(x_i + 1))
    plt.legend()
plt.subplot(515)
plt.plot(trace['error'], color = 'red', label = 'Error (RMSE)')
plt.ylabel('mm/day')
plt.legend()
plt.show()
```

Let's see how well the simulated streamflow fits the observed streamflow.

```python
q_mod = gr4j(x)
q_mod.set_conf(1)
q_sim = q_mod.run(peq)

plt.close('all')
plt.figure(figsize = (20, 10))
peq.q.plot(label = 'Observed streamflow')
q_sim.plot(label = 'Simulated streamflow')
plt.ylabel('mm/day')
plt.legend()
plt.show()
```
